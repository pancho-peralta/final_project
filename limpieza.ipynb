```python
def calcular_completitud(df_america):
    completitud = pd.DataFrame(df_america.isnull().sum()).reset_index()
    completitud = completitud.rename(columns={"index": "columna", 0: "faltantes"})
    completitud["completitud"] = (1 - completitud["faltantes"] / df_america.shape[0]) * 100
    completitud = completitud.sort_values(by="completitud", ascending=True)
    completitud.reset_index(drop=True, inplace=True)
    return completitud.round(0)
```
